```python
import numpy as np
import pandas as pd 
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('../data/snack_consumer_sensory_dataset.csv')
data.head()
```

```python


# Prepare data
sensory_vars = ['sweetness', 'crunchiness', 'oiliness', 'aftertaste']
X = StandardScaler().fit_transform(data[sensory_vars])

# PCA
pca = PCA()
scores = pca.fit_transform(X)

# Results
print("VARIANCE EXPLAINED:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {var:.1%}")

total_2pc = sum(pca.explained_variance_ratio_[:2])
print(f"\nFirst 2 PCs capture: {total_2pc:.1%} of sensory variation")

# Loadings
loadings = pd.DataFrame(
    pca.components_[:2].T,
    index=sensory_vars,
    columns=['PC1', 'PC2']
)
print("\nCOMPONENT LOADINGS:")
print(loadings.round(3))
```
